```python
numerical_columns = df.select_dtypes(include=['number']).columns
for col in numerical_columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in {col}: {len(outliers)}")
```

```python
plt.figure(figsize=(10, 6))
df.boxplot()
plt.title('Box Plot of Numerical Columns')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

* `Order ID:` The range of values is quite large (up to 1.0e9), suggesting a wide distribution of order IDs.
* `Units Sold, Unit Price, Unit Cost:` The ranges are relatively small, and there are no significant outliers in these columns.
* `Total Revenue, Total Cost, Total Profit:` These columns show some outliers. You identified 3 outliers in "Total Revenue", 3 in "Total Cost", and 5 in "Total Profit".

#### Feature Engineering

```python
df['Revenue per Unit'] = df['Total Revenue'] / df['Units Sold']

df['Cost per Unit'] = df['Total Cost'] / df['Units Sold']

df['Profit per Unit'] = df['Total Profit'] / df['Units Sold']
```

```python
df[['Revenue per Unit', 'Cost per Unit', 'Profit per Unit']].describe()
```

```python
df.dtypes
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.histplot(df['Revenue per Unit'], bins=30, kde=True)
plt.title('Distribution of Revenue per Unit')

plt.subplot(1, 3, 2)
sns.histplot(df['Cost per Unit'], bins=30, kde=True)
plt.title('Distribution of Cost per Unit')

plt.subplot(1, 3, 3)
sns.histplot(df['Profit per Unit'], bins=30, kde=True)
plt.title('Distribution of Profit per Unit')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
sns.scatterplot(x=df['Units Sold'], y=df['Revenue per Unit'])
plt.title('Units Sold vs Revenue per Unit')

plt.subplot(1, 3, 2)
sns.scatterplot(x=df['Units Sold'], y=df['Cost per Unit'])
plt.title('Units Sold vs Cost per Unit')

plt.subplot(1, 3, 3)
sns.scatterplot(x=df['Units Sold'], y=df['Profit per Unit'])
plt.title('Units Sold vs Profit per Unit')

plt.tight_layout()
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

df['Order Year'] = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day'] = df['Order Date'].dt.day
df['Order DayOfWeek'] = df['Order Date'].dt.dayofweek

df['Ship Year'] = df['Ship Date'].dt.year
df['Ship Month'] = df['Ship Date'].dt.month
df['Ship Day'] = df['Ship Date'].dt.day
df['Ship DayOfWeek'] = df['Ship Date'].dt.dayofweek
```

```python
csv_filename = 'Clean_Amazon_Sales_Data.csv'
df.to_csv(csv_filename, index = False)

from IPython.display import FileLink, display

display(FileLink(csv_filename))
```
